```python
class PositionCrawlerPattern(PositionCrawler):
    
    def balance_nodes(self, qtd):
    
        for nl in range(self._step):
            last = max(self._grid[nl], key=int)

            for np in range(qtd):
                posy = last + np + 1
                self._gpos.append_dummy((nl, posy))

    def col1_greater_col2(self, start_y, nposy):
        nstep = self._step+1
        
        diff = start_y - nposy
        for i in range(diff):
            posy = nposy + i
            self._gpos.append_dummy((nstep, posy))

        return diff

    def col1_lower_col2(self, start_y, nposy):
        diff = nposy - start_y
        for i in range(diff):
            posy = diff + i + start_y
            self._gpos.append_dummy((self._step, posy))

        return diff

    def grow_node(self, total, start):
        for ps in range(total-1):
            nps = (self._step, start + (ps+1))
            self._gpos.append_dummy(nps)
            self._gpos.inc_size()
        
    def soft_balance(self):
        succers = self._graph[self._node['label']]
        succers_size = len(succers)
        
        start_y = self._default_y()
        def_y = start_y

        nstep = self._step+1
        nposy = self._default_y(nstep)
        
        if start_y > nposy:
            diff = self.col1_greater_col2(start_y, nposy)
                
        if start_y < nposy:
            diff = self.col1_lower_col2(start_y, nposy)
            def_y = nposy + diff
        
        if succers_size >= 2:
            self.balance_nodes(succers_size-1)
            self.grow_node(succers_size, def_y)
            
        self._gpos.set_position((self._default_x(), def_y))
```
